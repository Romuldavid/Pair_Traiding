Частотный анализ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read data from file
data = pd.read_csv("ittensive.time.series.01.csv", decimal=',')
#change index
data = data[::-1].reset_index()
data.head()
```

```python
#расчитаем основные параметры распределения значений ряда
#draw curve
plt.figure(figsize=(12, 6))
plt.plot(data["Дата"].values, data["Цена"].values)
plt.xticks(data.index[::3], data["Дата"].values[::3], rotation=90)
plt.show()
```

```python
data["Цена"].describe()
```

```python
plt.figure(figsize=(12,6))
data["Цена"].hist(bins=20)
plt.show()
```

Оценим характер остатков от бегущего среднего

```python
data["Цена_6"] = data["Цена"].rolling(6).mean()
data["Цена_6_ост"] = data["Цена"] - data["Цена_6"]
```

```python
plt.figure(figsize=(12,6))
data["Цена_6_ост"].hist(bins=20)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data["Дата"].values, data["Цена_6_ост"].values)
plt.xticks(data.index[::3], data["Дата"].values[::3], rotation=90)
plt.show()
```

"Красный" шум ошибок

```python
plt.figure(figsize=(12,6))
np.abs(data["Цена_6_ост"]).hist(bins=20)
plt.show()
```
