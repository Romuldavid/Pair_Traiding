```python
#Расчет повторов
Stock = "RELIANCE.NS"
data = yf.download(Stock, start="2016-01-01", end="2023-02-14")
```

```python
#computation of returns
Stock = 'RELIANCE.NS'
data = yf.download(Stock, start="2016-01-01", end="2023-02-14")
T3= pd.DataFrame({"Close":data["Close"]})
T3['Year'] = T3.index.year
T2 = pd.DataFrame({"cumpnl_long":['0'], "cumpnl_short":['0'], "cumpnl":['0'], "SMA":['0'], "LMA":['0'], "Stock":['0'], "Year":['0']})
for z in range (2016, 2024, 1):
  T = T3.where(T3.Year == z)
  T = T.dropna()
  for x in range(1,35,2):
    for y in range(x,35,2):
      SMA=x
      LMA=y
      # Compute Moving averages of last 10 days and 30 days closing prices
      T['Short_average'] = T['Close'].rolling(window=SMA, min_periods=1,center=False).mean()
      T['Long_average'] = T['Close'].rolling(window=LMA, min_periods=1,center=False).mean()
      T['long_entry'] = T.Short_average > T.Long_average
      T['long_exit'] = T.Short_average <= T.Long_average
      T['positions_long'] = np.nan
      T.loc[T.long_entry,'positions_long'] = 1
      T.loc[T.long_exit,'positions_long'] = 0
      T.positions_long = T.positions_long.fillna(method='ffill')
      T['short_entry'] = T.Short_average < T.Long_average
      T['short_exit'] = T.Short_average >= T.Long_average
      T['positions_short'] = np.nan
      T.loc[T.short_entry,'positions_short'] = -1
      T.loc[T.short_exit,'positions_short'] = 0
      T.positions_short = T.positions_short.fillna(method='ffill')
      T['price_difference']= T.Close - T.Close.shift(1)
      T['pnllong'] = T.positions_long.shift(1) * T.price_difference
      T['pnlshort'] = T.positions_short.shift(1) * T.price_difference
      T['pnl'] = T['pnllong']+T['pnlshort']
      T['cumpnl_long'] = T.pnllong.cumsum()
      T['cumpnl_short'] = T.pnlshort.cumsum()
      T['cumpnl'] = T.pnl.cumsum()
      T1 = T[['cumpnl_short', 'cumpnl_long', 'cumpnl']].tail(1)
      T1['SMA'] = SMA
      T1['LMA'] = LMA
      T1['Stock'] = Stock
      T1['Year'] = z
      T2 = T2.append(T1)
print(T2)
# Pivot Table created
Pivot_Table1 = pd.pivot_table(T2, values ='cumpnl', index =['SMA', 'LMA'],
columns =['Year'], aggfunc = np.sum)
print (Pivot_Table1)

```

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
```

```python
Stock = "RELIANCE.NS"
data = yf.download(Stock, start="2016-01-01", end="2020-10-31")
```

```python
T3 = pd.DataFrame({"Close": data["Close"]})
T3['Year'] = T3.index.year
T2 = pd.DataFrame({"cumpnl":['0'], "MA":['0'], "STD":['0'], "Stock":['0'],
"Year":['0']})
for z in range (2016, 2021, 1):
    T = T3.where(T3.Year == z)
    T = T.dropna()
    for x in range(1,37,2):
        for y in range(1,3,1):
            MA=x
            STD=y
            T['moving_average'] = T.Close.rolling(MA).mean()
            T['moving_std_dev'] = T.Close.rolling(MA).std()
            T['upper_band'] = T.moving_average + (T.moving_std_dev*STD)
            T['lower_band'] = T.moving_average - (T.moving_std_dev*STD)
            T['long_entry'] = T.Close < T.lower_band
            T['long_exit'] = T.Close >= T.moving_average
            T['positions_long'] = np.nan
            T.loc[T.long_entry,'positions_long'] = 1
            T.loc[T.long_exit,'positions_long'] = 0
            T.positions_long = T.positions_long.fillna(method='ffill')
            T['short_entry'] = T.Close > T.upper_band
            T['short_exit'] = T.Close <= T.moving_average
            T['positions_short'] = np.nan
            T.loc[T.short_entry,'positions_short'] = -1
            T.loc[T.short_exit,'positions_short'] = 0
            T.positions_short = T.positions_short.fillna(method='ffill')
            T['positions'] = T.positions_long + T.positions_short
            T['price_difference']= T.Close - T.Close.shift(1)
            T['pnl'] = T.positions.shift(1) * T.price_difference
            T['cumpnl'] = T.pnl.cumsum()
            T1 = T[['cumpnl']].tail(1)
            T1['MA'] = MA
            T1['STD'] = STD
            T1['Stock'] = Stock
            T1['Year'] = z
            T2 = T2.append(T1)
print(T2)
```

```python
Pivot_Table1 = pd.pivot_table(T2, values ='cumpnl', index =['MA', 'STD'],
columns =['Year'], aggfunc = np.sum)
print (Pivot_Table1)
```

```python
Pivot_Table1.to_csv("PV_T.csv", index=True, encoding='utf8')
```

Мы берем цену закрытия HDFC и банка HDFC за последние 5 лет в
отдельной таблице 'prices_df' для расчета доходности по прошлым
данным. С помощью следующих кодов Python вы можете наблюдать за
изменением разницы цен обеих акций вместе с ценами обеих акций.

```python
prices_df = pd.DataFrame({"Stock_1": data["Close"], "Stock_2" : data1["Close"]})
prices_df['Diff'] = prices_df['Stock_1'] - prices_df['Stock_2']
prices_df[['Stock_1', 'Stock_2', 'Diff']].plot(grid=True, secondary_y= 'Diff', figsize=(14,6))
```

Теперь мы предпримем следующие шаги для расчета доходности
1. Сначала мы вычислим скользящее среднее разницы цен обеих акций. Как мы
обсуждали ранее, лучше использовать период скользящей средней как можно
короче. Итак, мы вычисляем 5-дневную скользящую среднюю.
2. Мы также вычисляем 5-дневное стандартное отклонение разницы в
цены обеих акций.
3. Как известно, 1 стандартное отклонение покрывает 66% значений. Это означает, что
если ценовая разница сдвинется более чем на 1 стандартное отклонение от
своего среднего значения, то будет высокая вероятность того, что цена
вернется к своему среднему значению.
4. Таким образом, мы вычисляем верхнюю полосу со средним значением + стандартное
отклонение и нижнюю полосу со средним значением - стандартное отклонение.
5. Мы создадим короткую позицию, когда разница цен пересечет верхнюю полосу, и
закроем позицию, когда она вернется к среднему значению. Мы создадим длинную
позицию, когда ценовая разница пересечет нижнюю полосу, и закроем позицию,
когда ценовая разница вернется к своему среднему значению.

```python
prices_df['moving_average'] = prices_df.Diff.rolling(5).mean()
prices_df['moving_std_dev'] = prices_df.Diff.rolling(5).std()
prices_df['upper_band'] = prices_df.moving_average + prices_df.moving_std_dev
prices_df['lower_band'] = prices_df.moving_average - prices_df.moving_std_dev
prices_df[['Diff', 'moving_average', 'upper_band', 'lower_band']].plot(figsize=(14,6))
```

```python
import plotly.express as px
```

```python
Table = prices_df[['Diff', 'moving_average', 'upper_band', 'lower_band']]
fig = px.line(Table)
fig.show()
```

```python
prices_df['long_entry'] = prices_df.Diff < prices_df.lower_band
prices_df['long_exit'] = prices_df.Diff >= prices_df.moving_average

prices_df['positions_long'] = np.nan
prices_df.loc[prices_df.long_entry,'positions_long'] = 1
prices_df.loc[prices_df.long_exit,'positions_long'] = 0
prices_df.positions_long = prices_df.positions_long.fillna(method='ffill')

prices_df['short_entry'] = prices_df.Diff > prices_df.upper_band
prices_df['short_exit'] = prices_df.Diff <= prices_df.moving_average
prices_df['positions_short'] = np.nan
prices_df.loc[prices_df.short_entry,'positions_short'] = -1
prices_df.loc[prices_df.short_exit,'positions_short'] = 0
prices_df.positions_short = prices_df.positions_short.fillna(method='ffill')

prices_df['positions'] = prices_df.positions_long + prices_df.positions_short
prices_df['price_difference']= prices_df.Diff - prices_df.Diff.shift(1)
prices_df['pnl'] = prices_df.positions.shift(1) * prices_df.price_difference
prices_df['cumpnl'] = prices_df.pnl.cumsum()

prices_df[['cumpnl']].plot(figsize=(16,8))
```

Большинство людей используют отношение «Z-оценка» (Акция 1 / Акция 2) для
генерации сигналов для торговли спредом. Однако я взял только разброс (разница в
цены) для всех расчетов. Если вы будете использовать «z-показатель», то результаты
будут почти такими же. Не произойдет, что расчет со спредом даст вам х прибыли, а расчет
с соотношением даст вам 2х прибыли, если данные и период одинаковы, то результаты
обоих останутся почти одинаковыми. Даже если вы сделаете некоторую оптимизацию,
создав более сложные инструменты для получения большей прибыли, эта оптимизация
будет проводиться на основе прошлых данных, и нет никакой гарантии, что
оптимизированные вещи будут работать лучше и в будущем. Сложность инструментов не
дает вам больше прибыли. Старайтесь, чтобы все было как можно проще. Когда вы
наблюдаете за ценой обеих акций, вы можете вычислить спред (разницу в цене) без
каких-либо инструментов или вычислений. У вас также есть представление о прошлых
спредах, например, мы видели разницу в цене HDFC и банка HDFC за последние 5 лет в
диапазоне от 600 до 1200. Мы торгуем с прибылью 50 или 100, поэтому, если этот спред идет
против нас, мы должны взять стоп-лосс, скажем, 100. Мы можем проверить, идет ли этот
спред в пользу или против, после создания позиции по спреду, просто наблюдая за ценами
в реальном времени. на экране. Таким образом, торговля со спредом проста для понимания
и не требует сложных вычислений.
Если вы хотите оптимизировать вещи для получения большей прибыли, оптимизируйте
дни. Сколько дней мы должны взять для вычисления среднего значения и стандартного
отклонения? Я взял 5 дней в приведенном выше примере, вы можете рассчитать прибыль
за 4 дня, 6 дней или может быть 10 дней. Одна вещь, которую вы заметите при тестировании
на прошлых данных: если вы возьмете больший период для среднего значения и
стандартного отклонения, то просадка будет большой. Стратегия даст вам большие
прибыли и большие убытки. Мы взяли период в 5 дней в приведенном выше примере, потому
что я заметил, что цены возвращаются к среднему на коротком периоде, однако стратегии
тренда на длинном периоде более прибыльны. Поэтому я считаю, что мы должны торговать
стратегиями возврата к среднему с самыми низкими из возможных краткосрочных
данных. Вы можете сформировать свое мнение на основе своего опыта; пробуйте все
больше и больше тестировать парную торговлю с различными комбинациями в разных

Мы
научились находить торговые возможности парной торговли
между двумя акциями одного сектора. Многие акции в одних
и тех же секторах доступны для торговли, мы можем
попытаться найти возможности парной торговли среди всех
акций, сравнивая акции один к одному. Следующие коды
можно использовать для расчета годовой прибыли от парной
торговли, полученной различными комбинациями частных
банков с 2016 по 2020 год. в среднем от 1 до 14.

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
from scipy.stats import pearsonr
from numpy import mean
from numpy import std
from statsmodels.tsa.stattools import coint, adfuller
import seaborn as sns
```

```python
# Получить исторические данные
data = yf.download('HDFC.NS', start="2016-01-01", end="2020-10-31")
data1 = yf.download('HDFCBANK.NS', start="2016-01-01", end="2020-10-31")
data2 = yf.download('ICICIBANK.NS', start="2016-01-01", end="2020-10-31")
data3 = yf.download('KOTAKBANK.NS', start="2016-01-01", end="2020-10-31")
data4 = yf.download('INDUSINDBK.NS', start="2016-01-01", end="2020-10-31")
data5 = yf.download('AXISBANK.NS', start="2016-01-01", end="2020-10-31")
```

```python
# Расчет корреляции, коинтеграция между всеми акциями
abc = ["HDFC", "HDFCBK", "ICICIBK", "KOTAKBK", "INDUSINDBK","AXISBK"]
prices_df = pd.DataFrame({"HDFC": data["Close"], "HDFCBK":
data1["Close"], "ICICIBK": data2["Close"], "KOTAKBK": data3["Close"],
"INDUSINDBK": data4["Close"],
"AXISBK": data5["Close"]})
```

```python
for aa in range(len(abc)):
    for bb in range(aa,len(abc)):
        if aa==bb: continue
        Cointegration = coint(prices_df[abc[aa]],prices_df[abc[bb]])[0]
        correlation = pearsonr(prices_df[abc[aa]],prices_df[abc[bb]])
        ADF = adfuller(prices_df[abc[aa]] - prices_df[abc[bb]], maxlag = 1)[1]
        print(abc[aa],abc[bb],"Cointegration",Cointegration,"correlation",correlation)
```

```python
#Вычисление годовой прибыли, генерируемой различными
prices_df['Year'] = prices_df.index.year
T2 = pd.DataFrame({"cumpnl":['0'], "Stock1":['0'], "Stock2":['0'], "Moving_average":['0'], "Year":['0']})
for z in range (2016, 2021, 1):
    T = prices_df.where(prices_df.Year == z)
    T = T.dropna()
    for x in range(len(abc)):
        for y in range(x,len(abc)):
            if x==y: continue
            for w in range(1,16,1):
                T['Diff'] = T[abc[x]]-T[abc[y]]
                T['moving_average'] = T.Diff.rolling(w).mean()
                T['moving_std_dev'] = T.Diff.rolling(w).std()
                T['upper_band'] = T.moving_average + T.moving_std_dev
                T['lower_band'] = T.moving_average - T.moving_std_dev
                T['long_entry'] = T.Diff < T.lower_band
                T['long_exit'] = T.Diff >= T.moving_average
                T['positions_long'] = np.nan
                T.loc[T.long_entry,'positions_long'] = 1
                T.loc[T.long_exit,'positions_long'] = 0
                T.positions_long = T.positions_long.fillna(method='ffill')
                T['short_entry'] = T.Diff > T.upper_band
                T['short_exit'] = T.Diff <= T.moving_average
                T['positions_short'] = np.nan
                T.loc[T.short_entry,'positions_short'] = -1
                T.loc[T.short_exit,'positions_short'] = 0
                T.positions_short = T.positions_short.fillna(method='ffill')
                T['positions'] = T.positions_long + T.positions_short
                T['price_difference']= T.Diff - T.Diff.shift(1)
                T['pnl'] = T.positions.shift(1) * T.price_difference
                T['cumpnl'] = T.pnl.cumsum()
                T1 = T[['cumpnl']].tail(1)
                T1['Stock1'] = abc[x]
                T1['Stock2'] = abc[y]
                T1['Moving_average'] = w
                T1['Year'] = z
                T2 = T2.append(T1)

print(T2)
```

```python
# Создана сводная таблица со значениями совокупной прибыли и убытков
Pivot_Table1 = pd.pivot_table(T2, values ='cumpnl', index =['Stock1',
'Stock2', 'Moving_average'], columns =['Year'], aggfunc = np.sum)
print (Pivot_Table1)
```
